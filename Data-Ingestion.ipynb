#Json

```python
import pandas as pd
from pyspark.sql import functions as F

url = "https://raw.githubusercontent.com/sharmadhiraj/free-json-datasets/refs/heads/master/docs/economy-finance/forbes_billionaires_list.json"

pdf = pd.read_json(url)

sdf = spark.createDataFrame(pdf)

target_path = "/Volumes/pyspark/bronze/raw_ingestion/json/forbes_billionaires_list"

(sdf
 .coalesce(1)
 .write
 .mode("overwrite")
 .json(target_path)
)

print("Written to:", target_path)
```

#Csv

```python
link = "https://drive.google.com/uc?id=1x2IdSNcHGLmot9i1h90gwMJr5lULC2QV&export=download"

df = pd.read_csv(link)

pdf = spark.createDataFrame(df)

target_path = "/Volumes/pyspark/bronze/raw_ingestion/csv/customers"

(
    pdf.coalesce(1)
    .write
    .mode("overwrite")
    .format("csv")
    .save(target_path)
)
```

# Parquet

```python
link = "https://www.timestored.com/data/sample/userdata.parquet"

df = pd.read_parquet(link)

sdf = spark.createDataFrame(df)
target_path = "/Volumes/pyspark/bronze/raw_ingestion/parquet/"
(
  sdf.coalesce(1)
  .write
  .mode("overwrite")
  .format("parquet")
  .save(target_path)
)
```
