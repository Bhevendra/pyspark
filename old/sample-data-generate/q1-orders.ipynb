```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_timestamp

spark = SparkSession.builder.getOrCreate()

data = [
    ("O1", "C1", "CREATED", 500.0, "2025-02-01 10:00:00"),
    ("O1", "C1", "SHIPPED", 500.0, "2025-02-02 12:00:00"),
    ("O1", "C1", "SHIPPED", 500.0, "2025-02-02 12:00:00"),  
    ("O2", "C2", "CREATED", 1500.0, "2025-02-01 09:00:00"),
    ("O2", "C2", "DELIVERED", 1500.0, "2025-02-03 14:00:00"),
    ("O3", "C3", "CREATED", 200.0, "2025-02-01 11:00:00"),
    ("O3", "C3", "CANCELLED", 200.0, "2025-02-02 08:00:00"),
    ("O3", "C3", "CANCELLED", 200.0, "2025-02-02 08:00:00") 
]

columns = ["order_id", "customer_id", "status", "amount", "updated_at"]

df = spark.createDataFrame(data, columns) \
          .withColumn("updated_at", to_timestamp(col("updated_at")))

df.show()
```

```python
df.write.mode("overwrite").saveAsTable("pyspark.sampledata.q1orders")
```
